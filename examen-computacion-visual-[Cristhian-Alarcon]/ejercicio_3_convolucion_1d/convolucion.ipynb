```python
# Ejercicio 3: Convolución Unidimensional en Python
# Implementación desde cero de la función de convolución 1D,
# generación de gráficas guardadas en 'graficos/' y mostradas inline

import os
import matplotlib.pyplot as plt

# 1. Definir señal de entrada (14 valores)
signal = [1, 3, 2, 5, 7, 6, 4, 8, 9, 2, 1, 0, 3, 5]

# 2. Definir kernel de tamaño 3
default_kernel = [1/3, 1/3, 1/3]
kernel = default_kernel

# 3. Implementar función de convolución sin numpy.convolve
def convolve_1d(x, h):
    N = len(x)
    M = len(h)
    y = [0] * (N + M - 1)
    for n in range(len(y)):
        acc = 0
        for k in range(M):
            if 0 <= n - k < N:
                acc += x[n - k] * h[k]
        y[n] = acc
    return y

# Ejecutar convolución
y = convolve_1d(signal, kernel)

# 4. Crear carpeta de gráficos
graphics_dir = 'graficos'
import os
os.makedirs(graphics_dir, exist_ok=True)

# Función auxiliar para guardar y mostrar gráficas
def save_and_show(path):
    plt.savefig(path)
    print(f"Guardado: {path}")
    plt.show()
    plt.close()

# Graficar señal original
plt.figure()
plt.stem(signal)
plt.title('Señal Original')
plt.xlabel('Índice n')
plt.ylabel('Amplitud')
signal_path = os.path.join(graphics_dir, 'señal_original.png')
save_and_show(signal_path)

# Graficar kernel
plt.figure()
plt.stem(kernel)
plt.title('Kernel')
plt.xlabel('Índice k')
plt.ylabel('Valor del Kernel')
kernel_path = os.path.join(graphics_dir, 'kernel.png')
save_and_show(kernel_path)

# Graficar resultado de la convolución
plt.figure()
plt.stem(y)
plt.title('Resultado de la Convolución 1D')
plt.xlabel('Índice n')
plt.ylabel('y[n]')
result_path = os.path.join(graphics_dir, 'resultado.png')
save_and_show(result_path)

print("Proceso completo: imágenes exportadas y mostradas y visualizadas inline.")
```
